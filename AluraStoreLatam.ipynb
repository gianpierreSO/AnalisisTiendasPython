### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 1. Análisis de facturación



```python
# Calcular el ingreso de cada tienda sumando la columna 'Precio'
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostrar los resultados formateados
print(f"Ingreso Total Tienda 1: {ingreso_tienda1:,.2f}")
print(f"Ingreso Total Tienda 2: {ingreso_tienda2:,.2f}")
print(f"Ingreso Total Tienda 3: {ingreso_tienda3:,.2f}")
print(f"Ingreso Total Tienda 4: {ingreso_tienda4:,.2f}")
```

```python
# Si también quieres ver el gran total de todas las tiendas juntas:
ingreso_total_global = ingreso_tienda1 + ingreso_tienda2 + ingreso_tienda3 + ingreso_tienda4
print(f"\nIngreso Total de todas las tiendas: {ingreso_total_global:,.2f}")
```

# 2. Ventas por categoría

```python
# Tienda 1
ventas_cat_tienda1 = tienda['Categoría del Producto'].value_counts()
print("--- Categorías más populares: Tienda 1 ---")
print(ventas_cat_tienda1)
print("\n")
```

```python
# Tienda 2
ventas_cat_tienda2 = tienda2['Categoría del Producto'].value_counts()
print("--- Categorías más populares: Tienda 2 ---")
print(ventas_cat_tienda2)
print("\n")
```

```python
# Tienda 3
ventas_cat_tienda3 = tienda3['Categoría del Producto'].value_counts()
print("--- Categorías más populares: Tienda 3 ---")
print(ventas_cat_tienda3)
print("\n")
```

```python
# Tienda 4
ventas_cat_tienda4 = tienda4['Categoría del Producto'].value_counts()
print("--- Categorías más populares: Tienda 4 ---")
print(ventas_cat_tienda4)
```

# 3. Calificación promedio de la tienda


```python
# Calcular el promedio de calificación para cada tienda
promedio_tienda1 = tienda['Calificación'].mean()
promedio_tienda2 = tienda2['Calificación'].mean()
promedio_tienda3 = tienda3['Calificación'].mean()
promedio_tienda4 = tienda4['Calificación'].mean()

print("--- Calificación Promedio por Tienda ---")
print(f"Tienda 1: {promedio_tienda1:.2f} estrellas")
print(f"Tienda 2: {promedio_tienda2:.2f} estrellas")
print(f"Tienda 3: {promedio_tienda3:.2f} estrellas")
print(f"Tienda 4: {promedio_tienda4:.2f} estrellas")
```

# 4. Productos más y menos vendidos

```python
# Agrupar las tiendas y sus nombres para procesarlas fácilmente
tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

for df, nombre in zip(tiendas, nombres_tiendas):
    # Contar cuántas veces se vendió cada producto
    conteo_productos = df['Producto'].value_counts()
    
    # Identificar el más y el menos vendido con sus cantidades
    mas_vendido = conteo_productos.index[0]
    cant_mas_vendido = conteo_productos.iloc[0]
    
    menos_vendido = conteo_productos.index[-1]
    cant_menos_vendido = conteo_productos.iloc[-1]
    
    # Imprimir el reporte de análisis
    print(f"==== REPORTE: {nombre} ====")
    print(f"Producto MÁS vendido: {mas_vendido} ({cant_mas_vendido} unidades)")
    print(f"Producto MENOS vendido: {menos_vendido} ({cant_menos_vendido} unidades)")
    print("\n" + "-"*30 + "\n")
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio para cada tienda
envio_promedio_t1 = tienda['Costo de envío'].mean()
envio_promedio_t2 = tienda2['Costo de envío'].mean()
envio_promedio_t3 = tienda3['Costo de envío'].mean()
envio_promedio_t4 = tienda4['Costo de envío'].mean()

print("--- Costo de Envío Promedio por Tienda ---")
print(f"Tienda 1: {envio_promedio_t1:,.2f}")
print(f"Tienda 2: {envio_promedio_t2:,.2f}")
print(f"Tienda 3: {envio_promedio_t3:,.2f}")
print(f"Tienda 4: {envio_promedio_t4:,.2f}")
```

# 6. VISUALIZACION

```python
import matplotlib.pyplot as plt
# Unir todas las tiendas agregando una columna para identificarlas
tienda['Sucursal'] = 'Tienda 1'
tienda2['Sucursal'] = 'Tienda 2'
tienda3['Sucursal'] = 'Tienda 3'
tienda4['Sucursal'] = 'Tienda 4'
df_completo = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
# Gráfico 1: Gráfico de Barras (Ingresos por Tienda
# Calcular ingresos
ingresos_por_tienda = df_completo.groupby('Sucursal')['Precio'].sum()

# Crear el gráfico
plt.figure(figsize=(8, 6))
barras = plt.bar(ingresos_por_tienda.index, ingresos_por_tienda.values, 
                 color=['#4C72B0', '#55A868', '#C44E52', '#8172B2'], edgecolor='black')

# Formato del gráfico
plt.title('Ingresos Totales por Tienda', fontsize=14, fontweight='bold')
plt.ylabel('Ingresos Totales ($)', fontsize=12)
plt.xlabel('Sucursal', fontsize=12)

# Agregar etiquetas de datos sobre las barras
plt.bar_label(barras, fmt='%d', padding=3, fontsize=10, fontweight='bold')
plt.ylim(0, ingresos_por_tienda.max() * 1.15) # Dar espacio para que no se corten los números

# Mostrar
plt.show()
```

```python
# Gráfico 2: Gráfico Circular o de Pastel (Distribución de Categorías)
# Contar la cantidad de ventas por categoría
categorias = df_completo['Categoría del Producto'].value_counts()

# Crear el gráfico
plt.figure(figsize=(8, 8))
plt.pie(categorias.values, labels=categorias.index, autopct='%1.1f%%', 
        startangle=90, colors=plt.cm.Pastel1.colors, 
        wedgeprops={'edgecolor': 'black', 'linewidth': 1})

# Formato del gráfico
plt.title('Distribución Global por Categoría de Producto', fontsize=14, fontweight='bold')

# Mostrar
plt.show()
```

```python
# Gráfico 3: Gráfico de Dispersión (Precio vs. Costo de Envío)

import matplotlib.pyplot as plt

# Crear el gráfico
plt.figure(figsize=(9, 6))
plt.scatter(df_completo['Precio'], df_completo['Costo de envío'], 
            alpha=0.6, color='teal', edgecolors='black', s=80)

# Formato del gráfico
plt.title('Relación: Precio del Producto vs Costo de Envío', fontsize=14, fontweight='bold')
plt.xlabel('Precio del Producto ($)', fontsize=12)
plt.ylabel('Costo de Envío ($)', fontsize=12)

# Agregar una cuadrícula para facilitar la lectura
plt.grid(True, linestyle='--', alpha=0.5)

# Mostrar
plt.show()
```

## 7.1. Introducción
El presente informe tiene como objetivo evaluar el rendimiento operativo, logístico y comercial de cuatro sucursales (Tienda 1, Tienda 2, Tienda 3 y Tienda 4). A través de un análisis exploratorio de datos, se han evaluado métricas clave como los ingresos totales, el volumen de ventas por categoría y producto, los costos de envío y la satisfacción del cliente. 

El propósito principal de este análisis es proporcionar una recomendación fundamentada en datos al Sr. Juan sobre **qué decisiones tomar con sus tiendas**, identificando cuál es la más fuerte para potenciar y cuál representa el eslabón más débil (candidata a reestructuración o venta/desinversión).

## 7.2. Desarrollo: Análisis de Resultados

### 7.2.1 Ingresos Totales
El análisis financiero revela el siguiente nivel de facturación para cada sucursal:
* **Tienda 1:** $1,150,880,400.00
* **Tienda 2:** $1,116,343,500.00
* **Tienda 3:** $1,098,019,600.00
* **Tienda 4:** $1,038,375,700.00

La **Tienda 1** lidera indiscutiblemente en facturación global, mientras que la **Tienda 4** presenta los ingresos más bajos del grupo, con una diferencia de más de 112 millones respecto a la tienda líder.

### 7.2.2 Desempeño de Productos y Categorías
La distribución general revela que la categoría **Muebles** es el principal motor de la empresa, siendo la más popular en absolutamente todas las sucursales. Sin embargo, al analizar el desempeño por producto, surgen particularidades:
* **El producto MÁS vendido en general** es el curso *"Iniciando en programación"* (65 unidades en la Tienda 2), seguido por *"Cama box"* (62 unidades en la Tienda 4) y *"Microondas"* (60 unidades en la Tienda 1).
* **Los productos con PEOR rendimiento** son *"Juego de mesa"* (32 unidades, Tienda 2), *"Celular ABXY"* (33 unidades, Tienda 1) y *"Guitarra eléctrica"* (33 unidades, Tienda 4).
* A pesar de que la Tienda 4 logró un alto volumen de ventas con su producto estrella (*Cama box*), su mezcla de productos no le permitió elevar su facturación total frente a las demás.

### 7.2.3 Satisfacción del Cliente (Calificaciones)
Al medir el nivel de satisfacción de los clientes en una escala de 1 a 5, encontramos los siguientes promedios:
* **Tienda 1:** 3.98 estrellas
* **Tienda 2:** 4.04 estrellas
* **Tienda 3:** 4.05 estrellas
* **Tienda 4:** 4.00 estrellas

Curiosamente, la **Tienda 1**, que es la que más dinero ingresa, es la única que cae por debajo de los 4.00 puntos, lo que sugiere que su alto nivel de ventas podría estar afectando su calidad de servicio o atención al cliente. La **Tienda 3** cuenta con los clientes más satisfechos (4.05).

### 7.2.4 Costos de Envío 
El costo de envío es un factor asumido por el cliente y varía según la tienda:
* **Tienda 1:** $26,018.61
* **Tienda 2:** $25,216.24
* **Tienda 3:** $24,805.68
* **Tienda 4:** $23,459.46

La **Tienda 1** transfiere a sus clientes el costo de envío más alto. No obstante, esto no ha impedido que sea la tienda que más ingresos genera, lo que indica que sus clientes están dispuestos a pagar la prima de envío. Por otro lado, la **Tienda 4** tiene los envíos más baratos, pero no logra convertir esto en un mayor volumen de ingresos.

## 7.3. Conclusión y Recomendación Estratégica

Basado en el análisis integral, si el objetivo del Sr. Juan es **vender (desinvertir en) una de las tiendas** para quedarse con las más fuertes, **la recomendación es vender la Tienda 4**. 

**Justificación:**
1. **Debilidad Financiera:** A pesar de tener los envíos más económicos y una calificación aceptable, es la sucursal que genera la menor cantidad de ingresos ($1,038,375,700.00). 
2. **Falta de Apalancamiento:** Sus bajos costos logísticos no se están traduciendo en ingresos competitivos ni en un aumento de ventas que compense la diferencia.

**Por otro lado, la Tienda 1 debe ser mantenida y priorizada**, ya que es el principal motor económico del negocio ($1.15 mil millones). Sin embargo, se le debe emitir una **alerta operativa**: es urgente revisar sus procesos de atención y despacho, ya que posee el envío más caro y la **calificación más baja (3.98)**. Si no se mejora la experiencia del cliente en la Tienda 1, los altos ingresos actuales podrían verse afectados a mediano plazo por la falta de fidelización.
